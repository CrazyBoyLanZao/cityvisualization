```python
import os
path = r"F:\Project\jiangxiLast\static\json\rainsheet" #文件夹目录
files= os.listdir(path) #得到文件夹下的所有文件名称
s = []
i = 0
for file in files: #遍历文件夹
     if not os.path.isdir(file): #判断是否是文件夹，不是文件夹才打开
        if(file == 'Untitled.ipynb'):
            continue
        data = pd.read_csv(file)
        if(i==0):
            df = data.groupby([data['城市名']]).mean()
            df = df.round({'AQI':0,'PM2.5':0,'PM10':0,'SO2':0,'NO2':0,'CO':0,'O3':0,'排名':0})
            df['日期'] = file.split('.')[0]+ '1日'
            i=1
        else:
            df1 = data.groupby([data['城市名']]).mean()
            df1 = df1.round({'AQI':0,'PM2.5':0,'PM10':0,'SO2':0,'NO2':0,'CO':0,'O3':0,'排名':0})
            df1['日期'] = file.split('.')[0] + '1日'
            df = df.append(df1)
```

```python
df.to_csv('month.csv',encoding='gbk')

```
